```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Download dos dados

```python
url_dados = 'https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv'

dataset = pd.read_csv(url_dados)

# explorar os dados de maneira inicial
dataset.head(8)
```

```python
dataset.Valor
```

## Limpeza dos dados

```python
# Maneira 1 de fazer a limpeza dos dados
# replaces concatenados, em cadeia

dataset.Valor.str.replace('R$ ', '', regex=False).str.replace('\n', '', regex=False).str.replace('/Ano', '').str.replace('/Dia', '').str.replace('/Mês', '').str.replace('.', '').str.strip().astype(int)
```

```python
## ou se quiser separar em partes
## recomendo fazer cada um em uma célula para ver o resultado passo a passo
replace1 = dataset.Valor.str.replace('R$ ', '', regex=False)
replace2 = replace1.str.replace('\n', '', regex=False)
replace3 = replace2.str.replace('/Ano', '').str.replace('/Dia', '').str.replace('/Mês', '')
replace_final = replace3.str.replace('.', '')
sem_espacos_extras = replace_final.str.strip()
final = sem_espacos_extras.astype(int)
final
```

## Salvando as novas colunas no dataframe antigo

```python
dataset[['Moeda', 'Valor_Anuncio', 'Tipo_Aluguel']] = dataset.Valor.str.split(expand=True)

###
### Cuidado para não usar somente valor para colocar o preço separado
### como ja existia uma coluna valor antes (com o preço com R$) o pandas
### tende a substituir a coluna antiga por valores novos (sem o R$)
### e o seu resultado será diferente do da aula
```

# Desafios

### 1 - Criar uma escala de valores em milhões

```python
somente_vendas_nova_escala = dataset[dataset['Tipo_Aluguel'].isnull()]
somente_vendas_nova_escala['Valor_Anuncio_Float'] = somente_vendas_nova_escala['Valor_Anuncio'].str.replace('.', '').astype(float)

somente_vendas_nova_escala.head(10)
```

```python
somente_vendas_nova_escala['Valor_Anuncio_Float'].apply(lambda x: x/1000000)
```

```python
somente_vendas_nova_escala['Valor_Anuncio_Float'] = somente_vendas_nova_escala['Valor_Anuncio_Float'].apply(lambda x: x/1000000)
```

```python
## SELECIONANDO COLUNAS ESPECIFICAS
pd.set_option('display.float_format', lambda x: '%.3f' % x)
somente_vendas_nova_escala[['Rua', 'Bairro', 'Cidade', 'Metragem', 'Quartos', 'Banheiros', 'Vagas', 'Valor_Anuncio_Float']].head(10)
```

```python
somente_vendas_nova_escala.describe()
```

### 2 - Melhorar a visualização do histograma

```python
somente_vendas_nova_escala['Valor_Anuncio_Float'].plot.hist()
```

```python
sns.set()
plt.figure(figsize=(17,7))
sns.color_palette("Paired")
plot = sns.histplot(data=somente_vendas_nova_escala, x='Valor_Anuncio_Float', kde=True, shrink=.8)

plt.suptitle('Histograma de Valor do Imóvel na Cidade de SP')
plt.title('Valores entre 0 e 15 milhões')
plt.xlabel('Valor do Imóvel')
plt.ylabel('Quantidade de Imóveis à Venda')

plot.legend(['Valores em escala de milhão'], prop={'size': 20})

plt.xlim((-1, 15))

plt.show()
```

### 3 - Calcular o m^2 por bairro

```python
somente_vendas_nova_escala.groupby('Bairro').mean()
```

```python
somente_vendas_nova_escala['Valor_m2'] = somente_vendas_nova_escala['Valor_Anuncio_Float']/somente_vendas_nova_escala['Metragem']
somente_vendas_nova_escala.head(6)
```

```python
somente_vendas_nova_escala.groupby('Bairro').mean()
```

```python
media_valores_bairro = somente_vendas_nova_escala.groupby('Bairro').mean()['Valor_Anuncio_Float']
media_metragem_bairro = somente_vendas_nova_escala.groupby('Bairro').mean()['Metragem']

## Preço médio do m2 por bairro na escala de milhoes
(media_valores_bairro/media_metragem_bairro)
```

```python
media_valores_bairro = somente_vendas_nova_escala.groupby('Bairro').mean()['Valor_Anuncio_Float']*1000
media_metragem_bairro = somente_vendas_nova_escala.groupby('Bairro').mean()['Metragem']

## Preço médio do m2 por bairro na escala de milhar
(media_valores_bairro/media_metragem_bairro)
```

```python
media_valores_bairro = somente_vendas_nova_escala.groupby('Bairro').mean()['Valor_Anuncio_Float']*1000
media_metragem_bairro = somente_vendas_nova_escala.groupby('Bairro').mean()['Metragem']

## Preço médio do m2 por bairro na escala de milhar
plt.figure(figsize=(17,8))
(media_valores_bairro/media_metragem_bairro).plot.line()
```

### 4 - Entender o gráfico abaixo

```python
media_valores_bairro = somente_vendas_nova_escala.groupby('Bairro').mean()['Valor_Anuncio_Float']*1000
media_metragem_bairro = somente_vendas_nova_escala.groupby('Bairro').mean()['Metragem']

## Preço médio do m2 por bairro na escala de milhar
plt.figure(figsize=(17,8))
plt.ylim((0, 15))  # para melhor ver os limites e desconsiderar outliers
(media_valores_bairro/media_metragem_bairro).plot.line()
```

Tirando exceções, os chamados outliers, possivelmente por erro nos dados, ou um valor realmente absurdo, a média de preços por metro quadrado dos imóveis por bairro na cidade de SP, varia mais entre 2.000 e  7.000 reais.
